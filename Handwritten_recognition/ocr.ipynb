```python
from transformers import pipeline

# Load pipeline with aggregation to get full entity spans like "Montair LC"
nlp = pipeline("ner", model="d4data/biomedical-ner-all", aggregation_strategy="simple")

text = "Patient takes Montair LC and Cidex daily"
entities = nlp(text)

# Filter only drug entities
medicines = [e['word'] for e in entities if e['entity_group'] == 'DRUG']
print("🧪 Extracted Medicines:", medicines)
```
